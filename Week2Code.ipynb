```python
import pandas as pd
import numpy as numpy

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
print("All good") # for conforming there is no error
```

#### Load the dataset

```python
df = pd.read_csv('PB_All_2000_2021.csv', sep=';')
df
```

```python
df.info()  
```

```python
df.shape 
```

#### Statistics of the data -> Mean, std deviation, min, max...

```python
df.describe()
```

#### Missing values

```python
df.isnull().sum()
```

#### Changing the date from object to date format

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

#### For checking columns and their data types (Here, we are checking whether the date format is changed or not)

```python
df.info() 
```

#### Sorting the values by id and date in ascending order

```python
df = df.sort_values(by=['id', 'date'])
df.head(20) # first 20 rows
```

#### Adding the columns with named year and month

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df
```

#### showing all the columns names in the dataset

```python
df.columns 
```

#### creating the list of pollutants with the chemical names

```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4', 'PO4', 'CL']
```

#### dropping the missing values using dropna() method

```python
df = df.dropna(subset=pollutants)
df.head()
```

#### Printing the no. of missing values in each column

```python
df.isnull().sum()
```

#### Feature and target selection -> Feature is an independent variable(Input) and Target is a dependent variable for Prediction

```python
X = df[['id', 'year']]
y = df[pollutants]
```

#### This line applies One-Hot Encoding to the id column in your X DataFrame

```python
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

#### Train, Test and Split -> X_train, X_test are inputs for training and testing where as y_train and y_test outputs

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

#### Training the model using using MultiOutputRegressor and RandomForestRegressor

```python
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

#### Evaluate model to make prediction

```python
y_pred = model.predict(X_test)
```

#### Loops over the model and print mean squared error and r square score

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

#### What will the pollutant levels be at the specific station in the specific year

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

#### Saving the trained models, so that we can use it later

```python
import joblib

joblib.dump(model, 'pollution_model.pkl')
joblib.dump(X_encoded.columns.tolist(), "model_columns.pkl")
print('Model and cols structure are saved!')
```
